# Sprint 03 – Análise Exploratória

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Configurações visuais

```python
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

### Carregando os dados

```python
df = pd.read_csv('br_rj_isp_estatisticas_seguranca_armas_apreendidas_mensal.csv')  
```

### 🔢 Análise de Variáveis Quantitativas

```python
# Selecionar variáveis numéricas
quantitativas = df.select_dtypes(include=np.number)

# Tabela resumo com medidas estatísticas
resumo_quantitativas = quantitativas.describe().T
resumo_quantitativas['moda'] = quantitativas.mode().iloc[0]
print("📊 Tabela resumo - Variáveis Quantitativas:")
display(resumo_quantitativas)

# Histograma de uma variável
quantitativas.hist(bins=20, figsize=(15, 10))
plt.suptitle("Distribuição das Variáveis Quantitativas")
plt.show()

# Boxplot de uma variável numérica
plt.figure()
sns.boxplot(data=quantitativas['quantidade_arma_fabricacao_caseira'])
plt.title('Boxplot: Quantidade de Armas Caseiras')
plt.show()
```

### 🔠 Análise de Variáveis Qualitativas

```python
# Selecionar variáveis não numéricas
qualitativas = df.select_dtypes(exclude=np.number)

# Tabela de frequência absoluta e relativa
for col in qualitativas.columns:
    print(f"\n📋 Variável: {col}")
    freq_abs = df[col].value_counts()
    freq_rel = df[col].value_counts(normalize=True) * 100
    tabela_freq = pd.DataFrame({'Frequência Absoluta': freq_abs, 'Frequência Relativa (%)': freq_rel})
    display(tabela_freq)

    # Gráfico de barras
    plt.figure()
    sns.countplot(x=col, data=df, order=freq_abs.index)
    plt.title(f'Gráfico de Barras: {col}')
    plt.xticks(rotation=45)
    plt.show()
```

### ❗ Diagnóstico dos Dados

```python
# Valores ausentes
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100
tabela_missing = pd.DataFrame({'Valores Nulos': missing, 'Porcentagem (%)': missing_percent})
print("\n🧼 Tabela de valores ausentes por variável:")
display(tabela_missing[tabela_missing['Valores Nulos'] > 0])

# Comentário sobre impacto (ajuste no seu relatório)
print("\nComentário: As variáveis com valores nulos devem ser analisadas individualmente para decidir entre preenchimento, remoção ou outra estratégia.")

# Duplicados
duplicados = df.duplicated().sum()
print(f"\n🔍 Linhas duplicadas: {duplicados}")
if duplicados > 0:
    print("⚠️ Existem linhas duplicadas. Você pode usar df.drop_duplicates() para removê-las se necessário.")
else:
    print("✅ Não há duplicatas no conjunto de dados.")
```

### 📊 Gráficos Adicionais com Interpretação

```python
# Gráfico de dispersão entre duas variáveis numéricas
if 'quant_apreendida' in quantitativas.columns and 'ano' in df.columns:
    plt.figure()
    sns.scatterplot(x='ano', y='quant_apreendida', data=df)
    plt.title('Dispersão: Armas Apreendidas por Ano')
    plt.show()
    print("📝 Interpretação: Esse gráfico mostra como o número de armas apreendidas evolui ao longo dos anos.")

# Boxplot por tipo de crime (exemplo com variável qualitativa)
if 'tipo_arma' in df.columns:
    plt.figure()
    sns.boxplot(x='tipo_arma', y='quant_apreendida', data=df)
    plt.title('Boxplot: Quantidade de Armas Apreendidas por Tipo')
    plt.xticks(rotation=45)
    plt.show()
    print("📝 Interpretação: Observa-se a distribuição das quantidades de apreensões por tipo de arma, destacando possíveis outliers.")
```
