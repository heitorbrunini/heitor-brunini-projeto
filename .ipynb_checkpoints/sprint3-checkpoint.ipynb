# Sprint 03 â€“ AnÃ¡lise ExploratÃ³ria

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### ConfiguraÃ§Ãµes visuais

```python
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

### Carregando os dados

```python
df = pd.read_csv('br_rj_isp_estatisticas_seguranca_armas_apreendidas_mensal.csv')  
```

### ðŸ”¢ AnÃ¡lise de VariÃ¡veis Quantitativas

```python
# Selecionar variÃ¡veis numÃ©ricas
quantitativas = df.select_dtypes(include=np.number)

# Tabela resumo com medidas estatÃ­sticas
resumo_quantitativas = quantitativas.describe().T
resumo_quantitativas['moda'] = quantitativas.mode().iloc[0]
print("ðŸ“Š Tabela resumo - VariÃ¡veis Quantitativas:")
display(resumo_quantitativas)

# Histograma de uma variÃ¡vel
quantitativas.hist(bins=20, figsize=(15, 10))
plt.suptitle("DistribuiÃ§Ã£o das VariÃ¡veis Quantitativas")
plt.show()

# Boxplot de uma variÃ¡vel numÃ©rica
plt.figure()
sns.boxplot(data=quantitativas['quantidade_arma_fabricacao_caseira'])
plt.title('Boxplot: Quantidade de Armas Caseiras')
plt.show()
```

### ðŸ”  AnÃ¡lise de VariÃ¡veis Qualitativas

```python
# Selecionar variÃ¡veis nÃ£o numÃ©ricas
qualitativas = df.select_dtypes(exclude=np.number)

# Tabela de frequÃªncia absoluta e relativa
for col in qualitativas.columns:
    print(f"\nðŸ“‹ VariÃ¡vel: {col}")
    freq_abs = df[col].value_counts()
    freq_rel = df[col].value_counts(normalize=True) * 100
    tabela_freq = pd.DataFrame({'FrequÃªncia Absoluta': freq_abs, 'FrequÃªncia Relativa (%)': freq_rel})
    display(tabela_freq)

    # GrÃ¡fico de barras
    plt.figure()
    sns.countplot(x=col, data=df, order=freq_abs.index)
    plt.title(f'GrÃ¡fico de Barras: {col}')
    plt.xticks(rotation=45)
    plt.show()
```

### â— DiagnÃ³stico dos Dados

```python
# Valores ausentes
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100
tabela_missing = pd.DataFrame({'Valores Nulos': missing, 'Porcentagem (%)': missing_percent})
print("\nðŸ§¼ Tabela de valores ausentes por variÃ¡vel:")
display(tabela_missing[tabela_missing['Valores Nulos'] > 0])

# ComentÃ¡rio sobre impacto (ajuste no seu relatÃ³rio)
print("\nComentÃ¡rio: As variÃ¡veis com valores nulos devem ser analisadas individualmente para decidir entre preenchimento, remoÃ§Ã£o ou outra estratÃ©gia.")

# Duplicados
duplicados = df.duplicated().sum()
print(f"\nðŸ” Linhas duplicadas: {duplicados}")
if duplicados > 0:
    print("âš ï¸ Existem linhas duplicadas. VocÃª pode usar df.drop_duplicates() para removÃª-las se necessÃ¡rio.")
else:
    print("âœ… NÃ£o hÃ¡ duplicatas no conjunto de dados.")
```

### ðŸ“Š GrÃ¡ficos Adicionais com InterpretaÃ§Ã£o

```python
# GrÃ¡fico de dispersÃ£o entre duas variÃ¡veis numÃ©ricas
if 'quant_apreendida' in quantitativas.columns and 'ano' in df.columns:
    plt.figure()
    sns.scatterplot(x='ano', y='quant_apreendida', data=df)
    plt.title('DispersÃ£o: Armas Apreendidas por Ano')
    plt.show()
    print("ðŸ“ InterpretaÃ§Ã£o: Esse grÃ¡fico mostra como o nÃºmero de armas apreendidas evolui ao longo dos anos.")

# Boxplot por tipo de crime (exemplo com variÃ¡vel qualitativa)
if 'tipo_arma' in df.columns:
    plt.figure()
    sns.boxplot(x='tipo_arma', y='quant_apreendida', data=df)
    plt.title('Boxplot: Quantidade de Armas Apreendidas por Tipo')
    plt.xticks(rotation=45)
    plt.show()
    print("ðŸ“ InterpretaÃ§Ã£o: Observa-se a distribuiÃ§Ã£o das quantidades de apreensÃµes por tipo de arma, destacando possÃ­veis outliers.")
```
